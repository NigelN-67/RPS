```python
import random

while True:
    arsenal = ["rock", "paper", "scissors"]

    rounds = int(input("You want 3 or 5 rounds? "))

    if rounds == 3:
        win_pts = 2
    elif rounds == 5:
        win_pts = 3
    else:
        print("Invalid choice. Default is 3 rounds.")
        rounds = 3
        win_pts = 2

    player_score = 0
    bot_score = 0
    round_number = 1

    while round_number <= rounds:

        print(f" Round {round_number} ---")

        You = input("Choose Wisely: rock, paper, or scissors? ").lower()
        
        if You not in arsenal:
            print("Pick from the three, please.")
            continue

        Bot = random.choice(arsenal)

        print("Bot chose:", Bot)

        if You == Bot:
            print("It's a tie!")

        elif You == "rock":
            if Bot == "scissors":
                print("Rock beats scissors; You WIN!")
                player_score += 1
            else:
                print("Paper beats rock... You LOSE!")
                bot_score += 1
        elif You == "paper":
            if Bot == "rock":
                print("Paper beats rock; You WIN!")
                player_score += 1
            else:
                print("Paper got scissored! You LOSE!")
                bot_score += 1
        elif You == "scissors":
            if Bot == "paper":
                print("Scissors beat paper; You WIN!")
                player_score += 1
            else:
                print("Scissors got stoned! You LOSE!")
                bot_score += 1

        print("Score â†’ You:", player_score, "| Computer:", bot_score)

        if player_score == win_pts or bot_score == win_pts:
            break

        round_number += 1

    print(" ***GAME OVER*** \n")

    if player_score > bot_score:
        print("YOU WON THE GAME!!!")
    elif bot_score > player_score:
        print("BOT WON THE GAME!!!")
    else:
        print("It's a tie!!!")

    one_more = input("Play Again? (y/n)")
    if one_more.lower() != "y":
        break
```
